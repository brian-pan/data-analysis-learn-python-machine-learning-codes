# Practice multiple different classification algorithms

### Models:

+ Linear Discriminant Analysis (LDA)
    + `from sklearn.discriminant_analysis import LinearDiscriminantAnalysis`
+ Logistic Regression (LR)
    + `from sklearn.linear_model import LogisticRegression`
+ Naive Bayes classifier
    + `from sklearn.naive_bayes import GaussianNB`
+ Support Vector Machines (SVM)
    + `from sklearn.svm import SVC`
+ k-Nearest Neighbors (kNN)
    + `from sklearn.neighbors import KNeighborsClassifier`
+ Decision Tree (DT)
    + `from sklearn.tree import DecisionTreeClassifier`


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x, y = make_classification(n_samples=5000,
                           n_features=2,
                           n_informative=2,
                           n_redundant=0,
                           n_classes=2,
                           n_clusters_per_class=1)

datasets = [make_moons(n_samples=500, noise=0.3),
            make_circles(n_samples=500, noise=0.2, factor=0.5),
            make_classification(n_samples=500,
                                n_features=2,
                                n_informative=2,
                                n_redundant=0,
                                n_classes=2,
                                n_clusters_per_class=1)]

plt.figure(figsize=(5, 15))
for i in range(len(datasets)):
    plt.subplot(311 + i)
    x, y = datasets[i]
    plt.scatter(x[:, 0], x[:, 1], c=y, s=30, cmap='Set1')
    plt.title(f"Data set {i + 1}")
```
