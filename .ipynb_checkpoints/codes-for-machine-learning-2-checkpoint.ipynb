```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# generate data 1,2,3
x, y = make_classification(n_samples=5000,
                           n_features=2,
                           n_informative=2,
                           n_redundant=0,
                           n_classes=2,
                           n_clusters_per_class=1)

datasets = [make_moons(n_samples=500, noise=0.3),
            make_circles(n_samples=500, noise=0.2, factor=0.5),
            make_classification(n_samples=500,
                                n_features=2,
                                n_informative=2,
                                n_redundant=0,
                                n_classes=2,
                                n_clusters_per_class=1)]

plt.figure(figsize=(5, 15))
for i in range(len(datasets)):
    plt.subplot(311 + i)
    x, y = datasets[i]
    plt.scatter(x[:, 0], x[:, 1], c=y, s=30, cmap='Set1')
    plt.title(f"Data set {i + 1}")
```

```python
# save data in dataset
x, y = datasets[1] # No.2 pic
# scatter plot
plt.scatter(x[:, 0], x[:, 1], c=y)  # x1x x2y c=color
plt.show()
```

```python
x.shape # 500 x/y
```

```python
#
x_train, x_test, y_train, y_test = train_test_split(x, y)
print(x_train.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = KNeighborsClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_test)
plt.title("Truth")
plt.subplot(1, 2, 2)
plt.scatter(x_test[:, 0], x_test[:, 1], c=y_pred)
plt.title("Predict")
plt.show()
```

```python
x_min = x[:, 0].min()
x_max = x[:, 0].max()
y_min = x[:, 1].min()
y_max = x[:, 1].max()


x_value = np.linspace(x_min, x_max, 100)
y_value = np.linspace(y_min, y_max, 100)

empty_container = np.zeros((x_value.shape[0] * y_value.shape[0], 2))
for i in range(x_value.shape[0]):
  empty_container[i * y_value.shape[0]:(i + 1) * y_value.shape[0], 0] = x_value[i]
  empty_container[i * y_value.shape[0]:(i + 1) * y_value.shape[0], 1] = y_value
```

```python
pred = model.predict(empty_container)
```

```python
plt.scatter(empty_container[:, 0], empty_container[:, 1], c=pred)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap="Set2")
plt.show()
```
